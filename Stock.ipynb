- Chỉ số EPS (tiếng Anh: Earnings Per Share) là lợi nhuận ròng (sau thuế) trên một cổ phiếu mà nhà đầu tư thu lại được.
  (khoản lời mà nhà đầu tư có được sau khi đã trừ đi vốn ban đầu bỏ ra để mua cổ phiếu đó)
- Chỉ số P/E (tiếng Anh: Price to Earning Ratio) là chỉ số đánh giá mối quan hệ giữa giá thị trường của một cổ phiếu (Price) và lợi nhuận ròng trên cổ phiếu đó (EPS).
  (giá nhà đầu tư sẵn sàng trả để mua cổ phiếu của một doanh nghiệp dựa theo lợi nhuận thu về của nó.)
- ROE (tiếng Anh: Return on common equyty) là tỷ suất lợi nhuận ròng trên vốn chủ sở hữu, đây được xem là tỷ số quan trọng nhất đối với cổ đông của công ty, bởi nó đong đo khả năng sinh lời trên mỗi đồng vốn của họ.
- ROA (tiếng Anh: Return on total assets) là thước đo khả năng sinh lời trên tổng tài sản của doanh nghiệp. Tài sản của doanh nghiệp được hình thành từ vốn chủ sở hữu và các nguồn vốn vay.
- Chỉ số P/B (tiếng Anh: Price to Book Ratio) là chỉ số dùng để so sánh giá của một cổ phiếu trên thị trường với giá trị sổ sách của cổ phiếu đó. Hiểu đơn giản là chỉ số này cho ta biết giá của cổ phiếu đang thấp hay cao hơn giá trên sổ sách tại doanh nghiệp.
- Chỉ số Beta (tiếng Anh: Beta coefficient) là chỉ số đo lường mức biến động, rủi ro hệ thống của một cổ phiếu hay một danh mục đầu tư trên thị trường chứng khoán.
- Chỉ số D/E (tiếng Anh là Debt to equity ratio) đây là chỉ số nợ trên vốn chủ sở hữu, là tỷ số giữa vốn nợ và vốn chủ sở hữu, nhằm xác định năng lực quản lý nợ và sức mạnh tài chính của doanh nghiệp.

```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def make_graph_revenue(list_stock_data, list_revenue_data, list_stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)
    for i in range(len(list_stock_data)):
      stock_data_specific = list_stock_data[i][list_stock_data[i].Date <= '2022-10-31']
      revenue_data_specific = list_revenue_data[i][list_revenue_data[i].Date <= '2022-10-31']
      fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date, infer_datetime_format=True), y=stock_data_specific.Close.astype("float"), name= list_stock[i]), row=1, col=1)
      fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date, infer_datetime_format=True), y=revenue_data_specific.Revenue.astype("float"), name= list_stock[i]), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title='All stock',
    xaxis_rangeslider_visible=True)
    fig.show()
```

```python
def make_graph_eps(eps_data):
    fig = make_subplots(rows=1, cols=1, shared_xaxes=True, subplot_titles=("Earnings Per Share"), vertical_spacing = .3)
    eps_data_specific = eps_data[eps_data.Date <= '2022-10-31']
    fig.add_trace(go.Bar(x=pd.to_datetime(eps_data_specific.Date, infer_datetime_format=True), y=eps_data_specific.EPS.astype("float"), name="EPS", marker=dict(color=eps_data_specific.EPS.astype("float"), coloraxis="coloraxis")))
    fig.update_xaxes(title_text="Date")
    fig.update_yaxes(title_text="EPS")
    fig.update_layout(coloraxis=dict(colorscale='Earth'), showlegend=False)
    fig.update_layout(showlegend=False,
    height=900,
    title='Earnings Per Share',
    xaxis_rangeslider_visible=True)
    fig.show()
```

```python
def make_graph_pe(pe_data):
    fig = make_subplots(rows=1, cols=1, shared_xaxes=True, subplot_titles=("Price to Earning Ratio"), vertical_spacing = .3)
    pe_data_specific = pe_data[pe_data.Date <= '2022-10-31']
    fig.add_trace(go.Scatter(x=pd.to_datetime(pe_data_specific.Date, infer_datetime_format=True), y=pe_data_specific.PE.astype("float"), name="PE Ratio"))
    fig.add_trace(go.Bar(x=pd.to_datetime(pe_data_specific.Date, infer_datetime_format=True), y=pe_data_specific.PE.astype("float"), name="PE", marker=dict(color=pe_data_specific.PE.astype("float"), coloraxis="coloraxis")))
    fig.update_xaxes(title_text="Date")
    fig.update_yaxes(title_text="PE Ratio")
    fig.update_layout(coloraxis=dict(colorscale='Picnic'), showlegend=False)
    fig.update_layout(showlegend=False,
    height=900,
    title= 'Price to Earning Ratio',
    xaxis_rangeslider_visible=True)
    fig.show()
```

```python
def make_graph_roe(roe_data):
    fig = make_subplots(rows=1, cols=1, shared_xaxes=True, subplot_titles=("Return on common equyty"), vertical_spacing = .3)
    roe_data_specific = roe_data[roe_data.Date <= '2022-10-31']
    fig.add_trace(go.Scatter(x=pd.to_datetime(roe_data_specific.Date, infer_datetime_format=True), y=roe_data_specific.ROE.astype("float"), name="ROE"))
    fig.add_trace(go.Bar(x=pd.to_datetime(roe_data_specific.Date, infer_datetime_format=True), y=roe_data_specific.ROE.astype("float"), name="ROE", marker=dict(color=roe_data_specific.ROE.astype("float"), coloraxis="coloraxis")))
    fig.update_xaxes(title_text="Date")
    fig.update_yaxes(title_text="ROE(%)")
    fig.update_layout(coloraxis=dict(colorscale='Portland'), showlegend=False)
    fig.update_layout(showlegend=False,
    height=900,
    title= 'Return on common equyty',
    xaxis_rangeslider_visible=True)
    fig.show()
```

- ttitle
- tname
- xtitle
- ytitleytitle

```python
def make_scatter_graph(data, ttitle, tname, xtitle, ytitle):
    fig = make_subplots(rows=1, cols=1, shared_xaxes=True, subplot_titles=ttitle, vertical_spacing = .3)
    data_specific = data[data.Date <= '2022-10-31']
    fig.add_trace(go.Scatter(x=pd.to_datetime(data_specific.Date, infer_datetime_format=True), y=data_specific.DATA.astype("float"), name= tname))
    fig.update_xaxes(title_text=xtitle)
    fig.update_yaxes(title_text=ytitle)
    fig.update_layout(showlegend=False,
    height=900,
    xaxis_rangeslider_visible=True)
    fig.show()
```

Đưa các chỉ số lên cùng 1 biểu đồ

```python
def total_graph_de(stock_data, revenue_data, eps_data, pe_data, roe_data, roa_data, de_data, stock):
  fig = make_subplots(rows=1, cols=1, shared_xaxes=True, subplot_titles='Total Data', vertical_spacing = .3)
  stock_data_specific = stock_data[stock_data.Date <= '2022-10-31']
  revenue_data_specific = revenue_data[revenue_data.Date <= '2022-10-31']
  eps_data_specific = eps_data[eps_data.Date <= '2022-10-31']
  pe_data_specific = pe_data[pe_data.Date <= '2022-10-31']
  roe_data_specific = roe_data[roe_data.Date <= '2022-10-31']
  roa_data_specific = roa_data[roa_data.Date <= '2022-10-31']
  de_data_specific = de_data[de_data.Date <= '2022-10-31']
  fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data_specific.Date, infer_datetime_format=True), y=stock_data_specific.Close.astype("float"), name="Share Price"))
  fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data_specific.Date, infer_datetime_format=True), y=revenue_data_specific.Revenue.astype("float"), name="Revenue"))
  fig.add_trace(go.Bar(x=pd.to_datetime(eps_data_specific.Date, infer_datetime_format=True), y=eps_data_specific.EPS.astype("float"), name="EPS", marker=dict(color=eps_data_specific.EPS.astype("float"), coloraxis="coloraxis")))
  fig.add_trace(go.Scatter(x=pd.to_datetime(pe_data_specific.Date, infer_datetime_format=True), y=pe_data_specific.PE.astype("float"), name="PE Ratio"))
  fig.add_trace(go.Scatter(x=pd.to_datetime(roe_data_specific.Date, infer_datetime_format=True), y=roe_data_specific.ROE.astype("float"), name="ROE"))
  fig.add_trace(go.Scatter(x=pd.to_datetime(roa_data_specific.Date, infer_datetime_format=True), y=roa_data_specific.ROA.astype("float"), name="ROA"))
  fig.add_trace(go.Scatter(x=pd.to_datetime(de_data_specific.Date, infer_datetime_format=True), y=de_data_specific.DE.astype("float"), name="D/E"))
  fig.update_xaxes(title_text="Date")
  fig.update_yaxes(title_text= stock)
  fig.update_layout(showlegend=False,
  height=900,
  xaxis_rangeslider_visible=True)
  fig.show()
```

```python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"")
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
```
